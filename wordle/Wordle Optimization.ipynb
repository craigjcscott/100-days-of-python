```python
import numpy as np
import pandas as pd
import wordle_answers
import wordle_guesses
```

```python
possible_answers = wordle_answers.possible_answers
allowed_guesses = wordle_guesses.allowed_guesses
```

```python
possible_answers[:10]
```

```python
allowed_guesses[:10]
```

```python
guess_dict = {}

for guess in allowed_guesses:
    
    overlap_total = []
    
    for answer in possible_answers:
        overlap = 0
        for letter in guess:
            if letter in answer:
                overlap += 1
        overlap_total.append(overlap)
    
    guess_dict[guess] = np.mean(overlap_total)
            
        
```

```python
guess_dict
```

```python
sorted(guess_dict, key=guess_dict.get, reverse=True)[:5]
```

```python
possible_answers_unique_letters = {}
unique_letters_list = []

for answer in possible_answers:
    unique_letters = []
    for letter in answer:
        if letter not in unique_letters:
            unique_letters.append(letter)
    unique_letters_list.append(unique_letters)
    
for word in unique_letters_list:
    for letter in word:
        if letter in possible_answers_unique_letters:
            possible_answers_unique_letters[letter] += 1
        else:
            possible_answers_unique_letters[letter] = 1
         
print(possible_answers_unique_letters)        
sorted(possible_answers_unique_letters, key=possible_answers_unique_letters.get, reverse=True)[:5]
```
